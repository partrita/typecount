# How much typing do you do in a day?

## Load the data

```python
import pandas as pd
import matplotlib.pyplot as plt
# from datetime import datetime
import seaborn as sns

df = pd.read_csv("../typing_count.csv")
# 날짜 열을 datetime 형식으로 변환
# df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

## Visualization

```python
# 그래프 그리기
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(df['Date'], df['Count'], marker='o')

# 윗쪽과 오른쪽 테두리 제거
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 그래프 꾸미기
ax.set_xlabel('', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

# x축 날짜 포맷 설정
fig.autofmt_xdate()

# y축 범위 설정 (0부터 시작하도록)
ax.set_ylim(bottom=0)

# 데이터 포인트에 값 표시
for i, count in enumerate(df['Count']):
    ax.annotate(str(count), (df['Date'][i], count), textcoords="offset points", xytext=(0,7), ha='center')

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python
# 전체 평균값 계산
overall_mean = df['Count'].mean()

# 스웜 플롯 그리기 (모든 Date를 하나의 그룹으로 묶음)
fig, ax = plt.subplots(figsize=(2, 3))

# 윗쪽과 오른쪽 테두리 제거
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

sns.swarmplot(x=['All Dates'] * len(df), y='Count', data=df, ax=ax)

# 평균값 표시
ax.axhline(y=overall_mean, color='red', linestyle='--', label=f'Mean: {overall_mean:.2f}')

# 그래프 꾸미기
ax.set_xlabel('')
ax.set_ylabel('Count', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

# 그래프 표시
plt.tight_layout()
plt.show()
print(overall_mean)
```
